```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
#Load Data

prices = pd.read_csv("City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv") #https://www.zillow.com/research/data/
rates = pd.read_csv("MORTGAGE30US.csv") #https://fred.stlouisfed.org/series/MORTGAGE30US

#Clean
prices['city_state'] = [f"{city} {state}"for city, state in zip(prices['RegionName'], prices['State'])]
```

# PRICES

```python
def plot_prices(df, city_state):
    
    city, state = city_state.split()
    x = df.columns[8:]
    y = df[(df['RegionName']==city) & (df['State']==state)][x].values[0]
    
    fig = go.Figure()
    fig.add_traces(go.Scatter(x=x, y=y, name = city_state))
    fig.update_layout(showlegend=True)
    return fig

def plot_multiple_prices(df, city_state_list):
    fig = go.Figure()
    for city_state in city_state_list:
        sub_fig = plot_prices(df, city_state)
        fig.add_traces(sub_fig.data[0])
    
    fig.update_xaxes(rangeslider_visible=True)
    fig.update_layout(showlegend=True, height=600)
    return fig

def plot_rates(df):
    fig = go.Figure()
    x = df['DATE']
    y = df['MORTGAGE30US']
    fig.add_traces(go.Scatter(x=x, y=y, name = "30-Year Fixed Rate Average"))
    
    fig.update_xaxes(rangeslider_visible=True)
    fig.update_layout(showlegend=True, height=600)
    return fig

def plot_prices_and_rates(prices, rates, city_state_list):
    fig = go.Figure()
    fig.add_traces(plot_rates(rates).data)
    fig.add_traces(plot_multiple_prices(prices, city_state_list).data)
    return fig

def plot_prices_and_rates(prices, rates, city_state_list):
    
    rates = rates[rates['DATE']>=prices.columns[8:].min()]
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    fig.add_trace(plot_rates(rates).data[0], secondary_y=True)
    for trace in plot_multiple_prices(prices, city_state_list).data:
        fig.add_trace(trace, secondary_y=False)
    fig.update_xaxes(rangeslider_visible=True)
    fig.update_layout(showlegend=True, height=600)
    return fig
```

# PERCENT CHANGE

```python
city_state_list = ["Houston TX", "Dallas TX"]

p = pd.DataFrame({"DATE" :prices.columns[8:-1]})
p["DATE"] = pd.to_datetime(p["DATE"]).dt.strftime('%m/%Y')
for city_state in city_state_list:
    p[city_state] = prices[prices["city_state"]==city_state][prices.columns[8:-1]].values[0]

r = rates[rates['DATE']>=prices.columns[8:].min()]
r.index = pd.to_datetime(r['DATE'])
r = r[['MORTGAGE30US']].resample('M').mean().reset_index()
r['DATE'] = r['DATE'].dt.strftime('%m/%Y')

merge = p.join(r.set_index('DATE'), on = 'DATE').dropna()
merge.index=merge['DATE']
merge.drop("DATE", axis = 1, inplace = True)
```

```python
periods = 120
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(go.Scatter(x = merge.index, 
                             y = merge['MORTGAGE30US'], 
                             name = "30-Year Fixed Rate Average")
                         , secondary_y=True)

for city_state in merge[city_state_list]:
    fig.add_trace(go.Scatter(x = merge.index, 
                             y = merge[city_state].pct_change(periods=periods)*100, 
                             name = city_state),
        secondary_y=False)
    
fig.update_xaxes(rangeslider_visible=True)
fig.update_layout(showlegend=True, height=600)
```
